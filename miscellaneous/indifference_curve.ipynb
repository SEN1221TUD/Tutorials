```python
# Import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Marginal utility of TC and TT (for linear case)
B_TC = -1.00
B_TT = -0.15
```

```python
# Alternative 1 is the reference alternative: 30 minutes and 5 euros
TT1 = 30
TC1 = 5

# Alternative 2 is the alternative which we want to be equally attractive (indifference)
TT2 = np.linspace(0, 60, 101)
TC2 = 10 # We are going to compute TC2. But, we give it a value here to ensure the code runs
```

### 1. Linear utility function

```python
# Linear utility functions
V1_lin = B_TC * TC1 + B_TT * TT1
V2_lin = B_TC * TC2 + B_TT * TT2
```

```python
# Indifference curves: V1_lin = V2_lin
# B_TC * TC1 + B_TT * TT1 = B_TC * TC2 + B_TT * TT2
# B_TC * TC1 - B_TC * TC2 = B_TT * TT2 - B_TT * TT1
# B_TC * (TC1 - TC2) = B_TT * (TT2 - TT1)
# TC1 - TC2 = B_TT / B_TC * (TT2 - TT1)
# TC2 = TC1 - B_TT / B_TC * (TT2 - TT1)
TC2_indiff = TC1 - B_TT / B_TC * (TT2 - TT1)
plt.plot(TT2, TC2_indiff, label='Indifference curve')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.plot(TT1, TC1, 'ro', label='Alt1')
plt.legend()
plt.show()
```

`Interpretation` **A person is indifferent between Alternative 1 (TT = 30, TC = €5), and alternative 2 with any combination of TT and TC that lies on the curve.**

### 2. Nonlinear utility function: diminishing and increasing marginal utility

#### 2.1 Diminishing marginal utility

```python
# Nonlinear utility functions where people have diminishing marginal utility for TT and linear marginal utility for TC
B_TTlog = -1.6 # Note that we use a larger B_TT to scale the plots and make the nonlinear utility functions more pronounced
V1_demin = B_TC * TC1 + B_TTlog * np.log(1+TT1)
V2_demin = B_TC * TC2 + B_TTlog * np.log(1+TT2)
```

#### 2.2 Increasing marginal utility

```python
# Nonlinear utility functions where people have increasing marginal utility for TT and linear marginal utility for TC
B_TTlin  = -0.001  # Note that we use smaller B_TT to scale the plots in a more realsitic domain (no negative TC)
B_TTsq = -0.003
V1_inc = B_TC * TC1 + B_TTlin * TT1 + B_TTsq * TT1**2
V2_inc = B_TC * TC2 + B_TTlin * TT2 + B_TTsq * TT2**2
```

```python
# Plot the utility as a function of TT
plt.plot(TT2, B_TT * TT2                     , label='V_TT linear utility')
plt.plot(TT2, B_TTlog * np.log(1+TT2)        , label='V_TT diminishing marginal utility')
plt.plot(TT2, B_TTlin * TT2 + B_TTsq * TT2**2, label='V_TT increasing marginal utility')
plt.xlabel('TT')
plt.ylabel('Utility')
plt.legend()
plt.title('Utility as a function of TT')
plt.show()
```

### 3. Indifference curves for nonlinear utility functions

#### 3.1 Indifference curve for decreasing marginal utility of TT

```python
# Indifference curves: V1_inc = V2_inc
# B_TC * TC1 + B_TTlog * log(1+TT1) = B_TC * TC2 + B_TTlog * log(1+TT2)
# B_TC * TC1 - B_TC * TC2 = B_TTlog * log(1+TT2) - B_TTlog * log(1+TT1)
# B_TC * (TC1 - TC2) = B_TTlog * log((1+TT2)/(1+TT1))
# TC1 - TC2 = B_TTlog / B_TC * log((1+TT2)/(1+TT1))
# TC2 = TC1 - B_TTlog / B_TC * log((1+TT2)/(1+TT1))
TC2_indiff = TC1 - B_TTlog / B_TC * np.log((1+TT2)/(1+TT1))
plt.plot(TT2, TC2_indiff, label='Indifference curve', color='C1')
plt.plot(TT1, TC1, 'ro', label='Alt1')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.legend()
plt.show()
```

#### 3.2 Indifference curve for increasing marginal utility of TT

```python
# Indifference curves: V1_inc = V2_inc
# B_TC * TC1 + B_TTlin * TT1 + B_TTsq * TT1**2 = B_TC * TC2 + B_TTlin * TT2 + B_TTsq * TT2**2
# B_TC * TC1 - B_TC * TC2 = B_TTlin * TT2 + B_TTsq * TT2**2 - B_TTlin * TT1 - B_TTsq * TT1**2
# B_TC * (TC1 - TC2) = B_TTlin * (TT2 - TT1) + B_TTsq * (TT2**2 - TT1**2)
# TC1 - TC2 = B_TTlin / B_TC * (TT2 - TT1) + B_TTsq / B_TC * (TT2**2 - TT1**2)
# TC2 = TC1 - B_TTlin / B_TC * (TT2 - TT1) - B_TTsq / B_TC * (TT2**2 - TT1**2)
TC2_indiff = TC1 - B_TTlin / B_TC * (TT2 - TT1) - B_TTsq / B_TC * (TT2**2 - TT1**2)
plt.plot(TT2, TC2_indiff, label='Indifference curve', color='C2')
plt.plot(TT1, TC1, 'ro', label='Alt1')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.legend()
plt.show()
```
