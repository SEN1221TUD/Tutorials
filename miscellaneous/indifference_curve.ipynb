```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
B_TC = -1.00
B_TT = -0.15
```

```python
# Alt1 is the reference alternative: 60 minutes and 10 euros
TC1 = 10
TT1 = 60

# Alt2 is the alternative which we want to be equally attractive (indifference)
TT2 = np.linspace(0, 60, 101)
TC2 = np.linspace(0, 10,  101)
```

### Linear utility function

```python
# Linear utility functions
V1_lin = B_TC * TC1 + B_TT * TT1
V2_lin = B_TC * TC2 + B_TT * TT2
```

```python
# Indifference curves: V1_lin = V2_lin
# B_TC * TC1 + B_TT * TT1 = B_TC * TC2 + B_TT * TT2
# B_TC * TC1 - B_TC * TC2 = B_TT * TT2 - B_TT * TT1
# B_TC * (TC1 - TC2) = B_TT * (TT2 - TT1)
# TC1 - TC2 = B_TT / B_TC * (TT2 - TT1)
# TC2 = TC1 - B_TT / B_TC * (TT2 - TT1)
TC2_indiff = TC1 - B_TT / B_TC * (TT2 - TT1)
plt.plot(TT2, TC2_indiff, label='Indifference curve')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.legend()
plt.show()
```

### Nonlinear utility function: increasing and decreasing marginal utility

```python
# Nonlinear utility functions where people have diminishing marginal utility for TT and linear marginal utility for TC
B_TTlog = -0.3 # Note that we use a larger B_TT to scale the plots and make the nonlinear utility functions more pronounced
V1_demin = B_TC * TC1 + B_TTlog * np.log(1+TT1)
V2_demin = B_TC * TC2 + B_TTlog * np.log(1+TT2)
```

```python
# Nonlinear utility functions where people have increasing marginal utility for TT and linear marginal utility for TC
B_TTlin  = -0.001  # Note that we use smaller B_TT to scale the plots in a more realsitic domain (no negative TC)
B_TTsq = -0.001
V1_inc = B_TC * TC1 + B_TTlin * TT1 + B_TTsq * TT1**2
V2_inc = B_TC * TC2 + B_TTlin * TT2 + B_TTsq * TT2**2
```

```python
# Plot the utility functions as a function of TT2
plt.plot(TT2, V2_lin, label='V2_TT linear')
plt.plot(TT2, V2_demin, label='V2_TT diminishing marginal utility')
plt.plot(TT2, V2_inc, label='V2_TT increasing marginal utility')
plt.xlabel('TT2')
plt.ylabel('Utility')
plt.legend()
plt.show()
```

#### Indifference for decreasing marginal utility of TT

```python
# Indifference curves: V1_inc = V2_inc
# B_TC * TC1 + B_TT * log(1+TT1) = B_TC * TC2 + B_TT * log(1+TT2)
# B_TC * TC1 - B_TC * TC2 = B_TT * log(1+TT2) - B_TT * log(1+TT1)
# B_TC * (TC1 - TC2) = B_TT * log((1+TT2)/(1+TT1))
# TC1 - TC2 = B_TT / B_TC * log((1+TT2)/(1+TT1))
# TC2 = TC1 - B_TT / B_TC * log((1+TT2)/(1+TT1))
TC2_indiff = TC1 - B_TT / B_TC * np.log((1+TT2)/(1+TT1))
plt.plot(TT2, TC2_indiff, label='Indifference curve')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.legend()
plt.show()
```

#### Indifference for increasing marginal utility of TT

```python
# Indifference curves: V1_inc = V2_inc
# B_TC * TC1 + B_TTlin * TT1 + B_TTsq * TT1**2 = B_TC * TC2 + B_TTlin * TT2 + B_TTsq * TT2**2
# B_TC * TC1 - B_TC * TC2 = B_TTlin * TT2 + B_TTsq * TT2**2 - B_TTlin * TT1 - B_TTsq * TT1**2
# B_TC * (TC1 - TC2) = B_TTlin * (TT2 - TT1) + B_TTsq * (TT2**2 - TT1**2)
# TC1 - TC2 = B_TTlin / B_TC * (TT2 - TT1) + B_TTsq / B_TC * (TT2**2 - TT1**2)
# TC2 = TC1 - B_TTlin / B_TC * (TT2 - TT1) - B_TTsq / B_TC * (TT2**2 - TT1**2)
TC2_indiff = TC1 - B_TTlin / B_TC * (TT2 - TT1) - B_TTsq / B_TC * (TT2**2 - TT1**2)
plt.plot(TT2, TC2_indiff, label='Indifference curve')
plt.xlabel('TT2')
plt.ylabel('TC2')
plt.legend()
plt.show()
```
